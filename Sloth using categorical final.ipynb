```python
#boxplot for claw_length_cm
sns.boxplot(x='claw_length_cm',data=df)
```

```python
#remove the outlier from claw_length
Q1 = df['claw_length_cm'].quantile(0.25)
Q3 = df['claw_length_cm'].quantile(0.75)
IQR = Q3-Q1
threshold=1.5 
admit_outliers=df[(df['claw_length_cm'] <Q1-threshold*IQR) | (df['claw_length_cm'] > Q3+ threshold*IQR)]
admit_outliers
```

```python
df=df.drop(admit_outliers.index)
sns.boxplot(x='claw_length_cm',data=df)
```

```python
#boxplot for size_cm
sns.boxplot(x='size_cm',data=df)
```

# Test and Train Splitting of x and y

```python
#spliting the data into x and y
x=df.iloc[:,0:4]
y=df.iloc[:,-1]
x
```

```python
#import train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=11)
x_train
```

```python
y_train
```

```python
#Transforming the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() #removes the mean and scales each feature/variable to unit variance.
x_train = scaler.fit_transform(x_train) 
x_test = scaler.transform(x_test)
x_test
```

# Logistic Regressor

```python
#import the logistic Regression library
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
#Assinging the parameter
from sklearn.model_selection import GridSearchCV
parameter ={'penalty':['l1','l2','elasticnet'],
             'C':[1,2,3,4,5,6,10,20,30,40,50],
              'max_iter':[100,200,300]}

classifier_regressor = GridSearchCV(classifier,
                                    param_grid = parameter,
                                    scoring='accuracy',
                                    cv = 5)

classifier_regressor.fit(x_train, y_train)
print(classifier_regressor.best_score_)
```

```python
#predicting x test
classifier_regressor.predict(x_test)
```

```python
#Assigning the y_pred
y_pred = classifier_regressor.predict(x_test)
y_pred
```

```python
#calculating the accuracy
from sklearn.metrics import accuracy_score,classification_report
score_lr = accuracy_score(y_pred,y_test)
print("Accuracy:",score_lr*100,"%")
```

```python
print(classification_report(y_pred,y_test))
```

```python
#predicting the result
lr=classifier_regressor.predict([[10.462,	9.705,	6.087	,1.308]])
lr
```

```python
lr2=classifier_regressor.predict([[8.662,51.498,4.551,1.259]])
lr2
```

# Random Forest

```python
# importing libraries for random forest
from sklearn.metrics import roc_curve,auc
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
import time
fit_rf = RandomForestClassifier(random_state = 40)
```

```python
#Hyperparameters
np.random.seed(40)
start=time.time()
param_dist={'max_depth':[2,3,4,5],
           'bootstrap':[True,False],
           'max_features':['auto','sqrt','log2',None],
           'criterion':['gini','entropy']}
cv_rf=GridSearchCV(fit_rf,cv=10,param_grid=param_dist,n_jobs=3)

cv_rf.fit(x_train,y_train)
print('best parameters using Grid Search: \n',cv_rf.best_params_)
end=time.time()
print('time taken in grid search: %0.2f'%(end-start) )
```
